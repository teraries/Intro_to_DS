```python
#загружаем библиотеки
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
#список названий столбцов
columns = ['Type', 'Alcohol','Malic acid','Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 
          'Proanthocyanins',  'Color intensity', 'Hue', 'OD280/OD315', 'Proline']
```

```python
#загружаем дату
data = pd.read_csv('wine.data', names = colums)
```

```python
#смотрим дату
data.head()
```

```python
#выделяем таргет
target = data['Type']
```

```python
#присваиваем таргет в вектор y
y = target
```

```python
# присваиваем матрицу признаков в Х
X = data[['Alcohol','Malic acid','Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 
          'Proanthocyanins',  'Color intensity', 'Hue', 'OD280/OD315', 'Proline']]
```

```python
data.head()
```

```python
#проводим перекрестную валидацию без нормироки
for i in range(1,50):
    model = KNeighborsClassifier(n_neighbors=i)
    kf = KFold(n_splits=5, shuffle=True, random_state = 42)
    res = cross_val_score(model, X, y, cv = kf, scoring = 'accuracy')
    print(np.mean(res))
```

```python
#лушчий результат: 1 сосед, 0,73
```

```python
# стандартизуем признаки
standardizer = StandardScaler()
X_std = standardizer.fit_transform(X)
```

```python
#проводим перекрестную валидацию на стандартизованных признаках
for i in range(1,50):
    model = KNeighborsClassifier(n_neighbors=i)
    kf = KFold(n_splits=5, shuffle=True, random_state = 42)
    res = cross_val_score(model, X_std, y, cv = kf, scoring = 'accuracy')
    print(np.mean(res))
```

```python
#лушчий результат: 29 соседей, 0.9776190476190475
```
